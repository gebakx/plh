## Tagger amb TextServer de FreeLing

```python
import requests
import json
```

```python
# set query elements
text = "L'Arnau té un gos. Se l'estima molt."
lang = 'ca'
out = 'json'
user = '???'
pwd = '???'
```

```python
# Create request
request_data = {'username':user,
                'password':pwd,
                'text_input':text,
                'language':lang,
                'output':out,
                'interactive':'1' }
```

```python
# service URL
service = "morpho"
url = "http://frodo.lsi.upc.edu:8080/TextWS/textservlet/ws/processQuery/"+service
```

```python
# Send request and get response    
resp = requests.post(url, files=request_data)
```

```python
# HTTP error, raise exception
if resp.status_code != requests.codes.ok : 
   resp.raise_for_status()
```

```python
# No error, appropriately process response
# (e.g. parsing XML or JSON, and doing clever stuff with the content)
print(resp.text)
```

```python
load = lambda r: json.loads(r.encode('utf-8'))
pars = lambda r: [p for p in r['paragraphs']]
sents = lambda p: [s for s in p['sentences']]
decode = lambda x: bytes(x,'latin1').decode('utf-8')
info = lambda s: [(decode(t['form']),t['tag'],t['pos']) for t in s['tokens']]
```

```python
list(map(info, sents(pars(load(resp.text))[0])))
```
